# Stack implementation using link list

```python
class ListNode:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def peek(self):
        if not self.isEmpty():
            x=self.top.data
            print("Peek Element: "+str(x))
        else:
            print("Stack is empty !")

    
    def isEmpty(self):
        if self.top == None:
            return True
        else:
            return False
        
    #traversing link list
    def traverse(self):
        if self.top is None:
            print("Stack is empty !")
            return
        else:
            p = self.top
            while p is not None :
                print(p.data)
                p = p.next
                
    #push element into stack
    def push(self,item):
        newNode = ListNode(item)
        newNode.next = self.top
        self.top = newNode
    
    #pop element from stack
    def pop(self):
        p = self.top
        prev = None
        #if link list is empty
        if self.top is None:
            print("Stack is empty ! cannot delete...")
            return 
        #if top is is to be deleted
        elif self.top.next == None:
            data = self.top.data
            self.top = None
            print(str(data)+" is deleted !")
            return 
        else:
            data = self.top.data
            self.top = self.top.next
            print(str(data)+" is deleted !")
            return

        
    #Searching element
    def search(self, item):
        if self.top is not None:
            p = self.top
            while p is not None and p.data != item  :
                p = p.next
            if p is None :
                print(str(item)+" is not present in stack !")
            else:
                print(str(item)+" is present in stack .")
        else:
            print("Stack is empty!")
            
    #sorting linklist
    def sort(self):
        if self.top is not None:
            p = self.top
            while p is not None:
                q = p.next
                while q is not None:
                    if q.data < p.data :
                        temp = p.data
                        p.data = q.data
                        q.data = temp
                    q = q.next
                p = p.next
        else:
            print("stack is empty!")
     
    #count number of nodes
    def count(self):
        cnt=0
        p=self.top
        while p is not None:
            cnt +=1
            p=p.next
        return cnt
    
```

```python
s1 = Stack() 

while True:
    print("-------------------------------------------------------------------------------------")
    print("\n                 Stack operations using link list ")

    print("1.Push")
    print("2.Pop")
    print("3.Peek element")
    print("4.Size(number of elements in stack)")
    print("5.Check Empty")
    print("6.Search")
    print("7.Sort")
    print("8.Traverse")
    print("9.Exit")
    choice= int(input("\n         Select operation : "))
    
    if choice == 1:
        x = int(input("\n Enter the data : "))
        s1.push(x)
                    
    elif choice == 2:
        s1.pop()
    
    elif choice == 3:
        s1.peek()
    
    elif choice == 4:
        x = s1.count()
        print("Size : "+str(x))
                
    elif choice == 5:
        if s1.isEmpty():
            print("Stack is empty")
        else:
            print("Stack is not empty ")
        
    elif choice == 6:
        x = int(input("\n Enter the data to search : "))
        s1.search(x)
        
    elif choice == 7:
        s1.sort()
        
    elif choice == 8:
        s1.traverse()
        
    elif choice == 9:
        print("\n Thank you !")
        break
                   
    else:
        print("\n Plaese select correct option !")
            
    
```
