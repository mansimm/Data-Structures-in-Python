# Complete implementation of queue

```python
class Queue:
    def __init__(self):
        self.que = []
        
    def __len__(self):
        return len(self.que)
        
    # Check Empty
    def isEmpty(self):
        if len(self.que)==0:
            return True
        else:
            return False
    
    # endqueue from rear
    def enque(self,data):
        self.que.append(data)
        
    # deque from front
    def deque(self):
        if self.isEmpty():
            print("Queue is empty , cannot delete !")
        else:
            x = self.que.pop(0)
            print("Deque element :"+str(x))
            
    #return peek element
    def peek(self):
        if self.isEmpty():
            print("Queue is empty.")
        else:
            print("Peek element : "+str(self.que[-1]))
            
    def search(self,y):
        for x in self.que:
            if x== y:
                print(str(y)+" is present in queue")
                return
        print(str(y)+" Not Found !")
        
    def traverse(self):
        print("Queue elements are :" )
        print(self.que)
    
```

```python
q1 = Queue()

while True:
    print("-------------------------------------------------------------------------------------")
    print("\n                 Queue operations using list ")

    print("1.Enque")
    print("2.Deque")
    print("3.Peek element")
    print("4.Size(number of elements in queue)")
    print("5.Check Empty")
    print("6.Search")
    print("7.Traverse")
    print("8.Exit")
    choice= int(input("\n         Select operation : "))
    
    if choice == 1:
        x = int(input("\n Enter the data : "))
        q1.enque(x)
                    
    elif choice == 2:
        q1.deque()
    
    elif choice == 3:
        q1.peek()
    
    elif choice == 4:
        x = len(q1)
        print("Size : "+str(x))
                
    elif choice == 5:
        if q1.isEmpty():
            print("Queue is empty")
        else:
            print("Queue is not empty ")
        
    elif choice == 6:
        x = int(input("\n Enter the data to search : "))
        q1.search(x)
        
    elif choice == 7:
        q1.traverse()
        
    elif choice == 8:
        print("\n Thank you !")
        break
                   
    else:
        print("\n Plaese select correct option !")
            
    
```
