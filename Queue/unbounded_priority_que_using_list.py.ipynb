# Unounded priority queue implementation using list

```python
class priorityQEntry(object):
    def __init__(self,item,priority):
        self.item = item
        self.priority = priority
```

```python
class priorityQ():
    def __init__(self):
        self.que = list()
        
    def __len__(self):
        return len(self.que)
    
    def isEmpty(self):
        return len(self) == 0
    
    #Priority is not bounded
    def enqueue(self,item,priority):
        entry = priorityQEntry(item,priority)
        self.que.append(entry)
        print(item+" is inserted .")
        self.sort()
            
    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty! cannot delete.")
        else:
            i=0
            highest = self.que.pop(0)
            print(str(highest.item)+" is deleted with highest priority "+str(highest.priority))
            
    def sort(self):
        if self.isEmpty():
            print("Queue is empty! cannot sort.")
        else:
            for i in range(len(self.que)):
                for j in range(len(self.que)-1):
                    if self.que[j].priority > self.que[j+1].priority:
                        temp = self.que[j]
                        self.que[j] = self.que[j+1]
                        self.que[j+1] = temp
            
    def peek(self):
        if self.isEmpty():
            print("Queue is empty!")
        else:
            print(" Peek Element is  "+self.que[0].item+" with highest priority of "+str(self.que[0].priority))
                  
    def search(self,y):
        for x in self.que:
            if x.item == y:
                print(str(y)+" is present in queue")
                return
        print(str(y)+" Not Found !")
        
    def traverse(self):
        print("Queue elements are :" )
        for x in self.que:
            print("Item : "+x.item+"  Priority : "+str(x.priority))
        
    
            
        
```

```python
q1 = priorityQ()

while True:
    print("-------------------------------------------------------------------------------------")
    print("\n           Unbounded Priority Queue operations using list ")

    print("1.Enque")
    print("2.Deque")
    print("3.Peek element(element with highest priority)")
    print("4.Size(number of elements in queue)")
    print("5.Check Empty")
    print("6.Search")
    print("7.Traverse")
    print("8.Exit")
    choice= int(input("\n         Select operation : "))
    
    if choice == 1:
        x = input("\n Enter the data (String) : ")
        y = int(input("\n Enter the priority : "))
        q1.enqueue(x,y)
                    
    elif choice == 2:
        q1.dequeue()
    
    elif choice == 3:
        q1.peek()
    
    elif choice == 4:
        x = len(q1)
        print(" Number of items in queue : "+str(x))
                
    elif choice == 5:
        if q1.isEmpty():
            print(" Queue is empty")
        else:
            print(" Queue is not empty ")
        
    elif choice == 6:
        x = str(input("\n Enter the data(item) to search : "))
        q1.search(x)
        
    elif choice == 7:
        q1.traverse()
        
    elif choice == 8:
        print("\n Thank you !")
        break
                   
    else:
        print("\n Plaese select correct option !")
            
    
```
