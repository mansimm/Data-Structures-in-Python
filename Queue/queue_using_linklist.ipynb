# Queue implementation using link list

```python
#defining link list node
class ListNode:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
#class queue
class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

        
    #check empty or not
    def isEmpty(self):
        if self.front==None and self.rear == None:
            return 1
        else:
            return 0
        
    #traversing link list
    def traverse(self):
        if self.front is None:
            print("Queue is empty !")
            return
        else:
            p = self.front
            while p is not None :
                print(p.data)
                p = p.next
                
    
    #insert element at the begining of list
    def enque(self,item):
        newNode = ListNode(item)
        newNode.next = self.front
        self.front = newNode
        if self.rear == None:
            self.rear = newNode
            
    
    #delete node from rear
    def deque(self):
        p = self.front
        prev = None
        #if link list is empty
        if self.front is None:
            print("Link list is empty ! cannot delete...")
            return 
        #if front and rear are poitiong to the same node
        elif self.front == self.rear:
            data = self.head.data
            self.head = None
            self.rear = None
            print(str(data)+" is deleted !")
            return 
        else:
            p = self.front
            while(p.next != self.rear):
                p=p.next
            data=p.next.data
            self.rear = p
            p.next = None
            print(str(data)+" is deleted !")
            return  

        
    #Searching element
    def search(self, item):
        if self.front is not None:
            p = self.front
            while p is not None and p.data != item  :
                p = p.next
            if p is None :
                print(str(item)+" is not present in queue !")
            else:
                print(str(item)+" is present in queue .")
        else:
            print("Queue is empty!")
            
    #sorting linklist
    def sort(self):
        if self.front is not None:
            p = self.front
            while p is not None:
                q = p.next
                while q is not None:
                    if q.data < p.data :
                        temp = p.data
                        p.data = q.data
                        q.data = temp
                    q = q.next
                p = p.next
        else:
            print("Queue is empty!")
     
    #count number of nodes
    def count(self):
        cnt=0
        p=self.front
        while p is not None:
            cnt +=1
            p=p.next
        return cnt
    
    #reverse link list 
    def reverse(self):
        if self.front is None:
            print("Sorry, Queue is empty !")
        else:
            prev = None
            current = self.front
            while current is not None:
                next = current.next
                current.next = prev
                prev = current
                current = next
            self.front = prev
         
    
                
   
                
```
