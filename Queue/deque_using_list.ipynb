```python
class DeQueue:
    def __init__(self):
        self.que = []
        
    def __len__(self):
        return len(self.que)
        
    # Check Empty
    def isEmpty(self):
        if len(self.que)==0:
            return True
        else:
            return False
    
    # endqueue from rear
    def enque_right(self,data):
        self.que.append(data)
        
    # endqueue from front
    def enque_left(self,data):
        self.que.insert(0,data)
        
    # deque from front
    def deque_left(self):
        if self.isEmpty():
            print("Queue is empty , cannot delete !")
        else:
            x = self.que.pop(0)
            print("Deque element :"+str(x))
            
    # deque from front
    def deque_right(self):
        if self.isEmpty():
            print("Queue is empty , cannot delete !")
        else:
            x = self.que.pop()
            print("Deque element :"+str(x))
            
    #return peek element
    def peek(self):
        if self.isEmpty():
            print("Queue is empty.")
        else:
            print("Front element : "+str(self.que[0]))
            print("Rear element : "+str(self.que[-1]))

            
    def search(self,y):
        for x in self.que:
            if x== y:
                print(str(y)+" is present in queue")
                return
        print(str(y)+" Not Found !")
        
    def traverse(self):
        print("Queue elements are :" )
        print(self.que)
    
```

```python
q1 = DeQueue()

while True:
    print("-------------------------------------------------------------------------------------")
    print("\n                 Deque operations using list ")

    print("1.Enque left")
    print("2.Enque right")
    print("3.Deque left")
    print("4.Deque right")
    print("5.Front and Rear element")
    print("6.Size(number of elements in dequeue)")
    print("7.Check Empty")
    print("8.Search")
    print("9.Traverse")
    print("10.Exit")
    choice= int(input("\n         Select operation : "))
    
    if choice == 1:
        x = int(input("\n Enter the data : "))
        q1.enque_left(x)
        
    elif choice == 2:
        x = int(input("\n Enter the data : "))
        q1.enque_right(x)
                    
    elif choice == 3:
        q1.deque_left()
        
    elif choice == 4:
        q1.deque_right()
    
    elif choice == 5:
        q1.peek()
    
    elif choice == 6:
        x = len(q1)
        print("Size : "+str(x))
                
    elif choice == 7:
        if q1.isEmpty():
            print("Dequeue is empty")
        else:
            print("Dequeue is not empty ")
        
    elif choice == 8:
        x = int(input("\n Enter the data to search : "))
        q1.search(x)
        
    elif choice == 9:
        q1.traverse()
        
    elif choice == 10:
        print("\n Thank you !")
        break
                   
    else:
        print("\n Plaese select correct option !")
            
    
```
