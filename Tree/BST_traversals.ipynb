```python
class Node:
    def __init__(self,data):
        self.left = None
        self.right = None
        self.data = data
```

```python
class Tree:
      
    #insert node in tree
    def insert(self,root,data):
        node = Node(data)
        
        if root is None:
            root = node
            return root
        
        else:
            if root.data < node.data:
                if root.right is None:
                    root.right = node
                else:
                    self.insert(root.right,data)
            else:
                if root.data > node.data:
                    if root.left is None:
                        root.left = node
                    else:
                        self.insert(root.left,data)                    
        return root
    
    
    #inorder traversal                    
    def inorder(self,root):
        if root is not None:
            self.inorder(root.left)
            print(root.data)
            self.inorder(root.right)
            
    #preorder traversal        
    def preorder(self,root):
        if root is not None:
            print(root.data)
            self.preorder(root.left)
            self.preorder(root.right)
       
    #postorder traversal
    def postorder(self,root):
        if root is not None:
            self.postorder(root.left)
            self.postorder(root.right)
            print(root.data)
            
    # check if tree is empty or not   
    def isEmpty(self,root):
        if root is None:
            return True
        else:
            return False
```

```python
T1 = Tree()
root = None

while True:
    print("-------------------------------------------------------------------------------------")
    print("\n                 Binary Search Tree ")

    print("1.Insert")
    print("2.Inorder traversal")
    print("3.Preorder Traversal")
    print("4.Postorder Traversal")
    print("5.Exit")
    choice= int(input("\n         Select operation : "))
    
    if choice == 1:
        x = int(input("\n Enter the data : "))
        root = T1.insert(root,x)
                    
    elif choice == 2:
        if not T1.isEmpty(root):
            T1.inorder(root)
        else:
            print("Tree is empty.") 
    
    elif choice == 3:
        if not T1.isEmpty(root):
            T1.preorder(root)
        else:
            print("Tree is empty.")      
    
    elif choice == 4:
        if not T1.isEmpty(root):
            T1.postorder(root)
        else:
            print("Tree is empty.") 
            
    elif choice == 5:
        print("\n Thank you !")
        break
               
                   
    else:
        print("\n Plaese select correct option !")
```
