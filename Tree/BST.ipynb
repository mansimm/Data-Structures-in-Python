# Binary Search Tree

```python
class Node:
    def __init__(self,data):
        self.left = None
        self.right = None
        self.data = data
```

```python
class Tree:
      
    #insert node in tree
    def insert(self,root,data):
        node = Node(data)
        
        if root is None:
            root = node
            return root
        
        else:
            if root.data < node.data:
                if root.right is None:
                    root.right = node
                else:
                    self.insert(root.right,data)
            else:
                if root.data > node.data:
                    if root.left is None:
                        root.left = node
                    else:
                        self.insert(root.left,data)                    
        return root
    
    
    #returns node with minimum value node    
    def minValNode(self,root):
        current = root
        while current.left is not None:
            current = current.left 
        return current
    
    
    #delete node from bst
    def delete(self,root,data):
        if root is not None:
            if root.data > data:
                root.left = self.delete(root.left,data)
                
            elif root.data < data:
                root.right = self.delete(root.right,data)
                
            else:
                if root.right is None:
                    temp = root.left
                    root = None
                    return temp
                
                elif root.left is None:
                    temp = root.right
                    root = None
                    return temp
                
                else:
                    min_node = self.minValNode(root.right)
                    root.data = min_node.data
                    root.right = self.delete(root.right,min_node.data)        
        return root
    
                        
    def inorder(self,root):
        if root is not None:
            self.inorder(root.left)
            print(root.data)
            self.inorder(root.right)
            
    def count(self,root):
        cnt = 1
        if root.left is not None :
            cnt += self.count(root.left)
        if root.right is not None:
            cnt += self.count(root.right)
        return cnt
       
    # search the node with given data
    def search(self,root,data):
        search = False
        if root is not None:
            if root.data == data:
                return True
            elif root.data > data:
                search = self.search(root.left,data)
            elif root.data < data:
                search = self.search(root.right,data)
        return search
            
        
    def height(self,root):
        height = 0
        if root is not None:
            height = 1
            
            height_left = self.height(root.left)
            height_right = self.height(root.right)
                
            if height_left > height_right:
                height += height_left
            else:
                height += height_right
                
        return height
        
    # check if tree is empty or not   
    def isEmpty(self,root):
        if root is None:
            return True
        else:
            return False
```
